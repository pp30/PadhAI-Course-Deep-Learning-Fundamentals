#Python classes

```
class mobilephones:
  '''This is a sample class to illustrate how python classes work '''
  def __init__(self,name,is_ios=False,screen_size=6.2):
    self.name=name
    self.is_ios = is_ios
    self.screen_size = screen_size
    self.rating = -1
  
  def has_rating(self):
    return self.rating > -1
```

```
new_phone=mobilephones('asus_maxpro')
```

```
type(new_phone)
```

```
print(new_phone.name,new_phone.is_ios,new_phone.screen_size)
```

```
new_phone.screen_size=4
```

```
new_phone.has_rating()
```

```
new_phone.rating=3.9
```

```
mobilephones.__doc__
```

```
class iphone(mobilephones):
  def __init__(self,name):
    mobilephones.__init__(self,name,True,4)
  
  def __str__(self):
    return self.name+ " "+ str(self.is_ios)+ " "+ str(self.screen_size)
    
```

```
naya = iphone('iphone11pro')
naya.is_ios
```

```
print(naya)
```
